```python

import pandas as pd
df = pd.read_csv('/2006new.csv')
df.head(10)
# summary = df.describe()
# print(summary)
```

```python
from tabulate import tabulate
```

```python
df.info()
```

```python
# Check for null values in each column and sum them up
null_counts = df.isnull().sum()

# Display the count of null values for each column
print(null_counts)
```

```python
df['DepTime'] = df['DepTime'].fillna(df['DepTime'].mean())
df['ArrTime'] = df['ArrTime'].fillna(df['ArrTime'].mean())
df['ActualElapsedTime'] = df['ActualElapsedTime'].fillna(df['ActualElapsedTime'].mean())
df['ArrDelay'] = df['ArrDelay'].fillna(df['ArrDelay'].mean())
df['DepDelay'] = df['DepDelay'].fillna(df['DepDelay'].mean())
df['Distance'] = df['Distance'].fillna(df['Distance'].mean())
```

```python
df['AirTime'] = df['AirTime'].fillna(df['AirTime'].mean())
```

```python
df['CRSElapsedTime'] = df['CRSElapsedTime'].fillna(df['CRSElapsedTime'].mean())
```

```python
df.drop('CancellationCode', axis=1, inplace=True)
```

```python
# Check for null values in each column and sum them up
null_counts = df.isnull().sum()

# Display the count of null values for each column
print(null_counts)
```

```python
df.info()
```

```python
sample_size = 30000
random_sample = df.sample(n=sample_size, random_state=42)
```

```python

random_sample['Delayed'] = random_sample.apply(lambda row: 'N' if row['ArrDelay'] <= 0 and row['DepDelay'] <= 0 else 'Y', axis=1)

```

```python
first_few_rows = random_sample.head(5)
print(tabulate(first_few_rows, headers='keys', tablefmt='pretty'))
```

```python
# Check for null values in each column and sum them up
null_counts = random_sample.isnull().sum()

# Display the count of null values for each column
print(null_counts)
```

```python
# Assuming 'df' is your original DataFrame
# Replace 'df' with 'random_sample' and fill NaN values with column means

random_sample['DepTime'] = random_sample['DepTime'].fillna(random_sample['DepTime'].mean())
random_sample['ArrTime'] = random_sample['ArrTime'].fillna(random_sample['ArrTime'].mean())
random_sample['ActualElapsedTime'] = random_sample['ActualElapsedTime'].fillna(random_sample['ActualElapsedTime'].mean())
random_sample['ArrDelay'] = random_sample['ArrDelay'].fillna(random_sample['ArrDelay'].mean())
random_sample['DepDelay'] = random_sample['DepDelay'].fillna(random_sample['DepDelay'].mean())
random_sample['Distance'] = random_sample['Distance'].fillna(random_sample['Distance'].mean())
random_sample['AirTime'] = random_sample['AirTime'].fillna(random_sample['AirTime'].mean())
random_sample['CRSElapsedTime'] = random_sample['CRSElapsedTime'].fillna(random_sample['CRSElapsedTime'].mean())
random_sample.drop('CancellationCode', axis=1, inplace=True)
```

```python
random_sample['TaxiOut'] = df['TaxiOut'].astype(int)
random_sample['TaxiIn'] = df['TaxiIn'].astype(int)
random_sample['CRSElapsedTime'] = df['CRSElapsedTime'].astype(int)
```
